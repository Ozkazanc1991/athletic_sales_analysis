### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

# Show the top 5 results.

import pandas as pd

file_2020 = "Resources/athletic_sales_2020.csv"
file_2021 = "Resources/athletic_sales_2021.csv"

# Reading the two CSV files into dataframes
df_2020 = pd.read_csv(file_2020)
df_2021 = pd.read_csv(file_2021)

# Code to Concatenate the two dataframes
combined_df = pd.concat([df_2020, df_2021])

# Grouping by region, state, and city and summing up the units sold
grouped_df = combined_df.groupby(['region', 'state', 'city']).agg({'units_sold': 'sum'}).reset_index()

# Renaming the sum column to "Total_Products_Sold"
grouped_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Code to Show the top 5 results
top_5_results = grouped_df.nlargest(5, 'Total_Products_Sold')

print(top_5_results)


                            
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"

# Show the top 5 results.

import pandas as pd

file_2020 = "Resources/athletic_sales_2020.csv"
file_2021 = "Resources/athletic_sales_2021.csv"

# Reading CSV files into dataframes
df_2020 = pd.read_csv(file_2020)
df_2021 = pd.read_csv(file_2021)

# Renaming the "total_sales" column to "Total Sales" for both dataframes
df_2020.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
df_2021.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Showing the total sales for the products sold for each retailer, region, state, and city
# Group by the desired columns and sum the 'Total Sales'
total_sales_2020 = df_2020.groupby(['retailer', 'region', 'state', 'city'])['Total Sales'].sum().reset_index()
total_sales_2021 = df_2021.groupby(['retailer', 'region', 'state', 'city'])['Total Sales'].sum().reset_index()

# Showing the top 5 results for each year
top_5_2020 = total_sales_2020.nlargest(5, 'Total Sales')
top_5_2021 = total_sales_2021.nlargest(5, 'Total Sales')

print("The Top 5 Total Sales for 2020:")
print(top_5_2020)
print("\nThe Top 5 Total Sales for 2021:")
print(top_5_2021)

```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.

import pandas as pd

file_2020 = "Resources/athletic_sales_2020.csv"
file_2021 = "Resources/athletic_sales_2021.csv"

# Reading CSV files into dataframes
df_2020 = pd.read_csv(file_2020)
df_2021 = pd.read_csv(file_2021)

# Rename the "total_sales" column to "Total Sales"
df_2020.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
df_2021.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Creating pivot table with 'invoice_date' as index and 'Total Sales' as values
pivot_table_2020 = pd.pivot_table(df_2020, index='invoice_date', values='Total Sales', aggfunc='sum')
pivot_table_2021 = pd.pivot_table(df_2021, index='invoice_date', values='Total Sales', aggfunc='sum')

# Showing the table of the results
print("Pivot Table for 2020:")
print(pivot_table_2020)

print("\nPivot Table for 2021:")
print(pivot_table_2021)

```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.


# Sort the resampled pivot table in ascending order on "Total Sales".

import pandas as pd

file_2020 = "Resources/athletic_sales_2020.csv"
file_2021 = "Resources/athletic_sales_2021.csv"

# Reading CSV files into dataframes
df_2020 = pd.read_csv(file_2020)
df_2021 = pd.read_csv(file_2021)

# Code for Converting 'invoice_date' column to datetime type
df_2020['invoice_date'] = pd.to_datetime(df_2020['invoice_date'])
df_2021['invoice_date'] = pd.to_datetime(df_2021['invoice_date'])

# Rename the "total_sales" column to "Total Sales"
df_2020.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
df_2021.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Creating pivot table with 'invoice_date' as index and 'Total Sales' as values
pivot_table_2020 = pd.pivot_table(df_2020, index='invoice_date', values='Total Sales', aggfunc='sum')
pivot_table_2021 = pd.pivot_table(df_2021, index='invoice_date', values='Total Sales', aggfunc='sum')

# Code to Resample the pivot table into daily bins and get the total sales for each day
resampled_pivot_2020 = pivot_table_2020.resample('D').sum()
resampled_pivot_2021 = pivot_table_2021.resample('D').sum()

# Sorting the resampled pivot table in ascending order based on "Total Sales"
sorted_resampled_2020 = resampled_pivot_2020.sort_values(by='Total Sales', ascending=True)
sorted_resampled_2021 = resampled_pivot_2021.sort_values(by='Total Sales', ascending=True)

print("Resampled and Sorted Pivot Table for 2020:")
print(sorted_resampled_2020)

print("\nResampled and Sorted Pivot Table for 2021:")
print(sorted_resampled_2021)


```
